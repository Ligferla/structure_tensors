```python
def create_csv_file_by_sample(sample, kind, poles_dir):
    
    path_to_csv = '/home/ligferla/' + kind + '_' + sample + '_poles_features.csv'
    with open(path_to_csv, 'w') as csvfile:
        fieldnames = ['img_id', 'm_0_0', 'm_0_1', 'm_1_1', 'lamda_max','lamda_min', 'min_intensity',
                      'max_intensity', 'mean', 'standard_deviation', 'lamda_max_min_diff', 'answer']
        writer = csv.DictWriter(csvfile, delimiter=",", fieldnames=fieldnames)
        writer.writeheader()
            
#         poles_dir = '/tmp/train/' + flag + '_poles/'
        for root, dirs, filenames in os.walk(poles_dir):
            for f in filenames:
                
                path_to_file = poles_dir + f
                input_image_path = "-input_image_path=" + path_to_file
                args = ('/home/ligferla/swarm1/bin.linux64.debug/structure_tensors_test', input_image_path)
                popen = subprocess.Popen(args, stdout=subprocess.PIPE)
                popen.wait()
                output = popen.stdout.read().split()
                answer = 0
                if (kind == 'positive'):
                    answer = 1 
                writer.writerow({'img_id': f,
                                 'm_0_0': output[0],
                                 'm_0_1': output[1],
                                 'm_1_1': output[2],
                                 'lamda_max': output[3],
                                 'lamda_min': output[4],
                                 'min_intensity' : output[5],
                                 'max_intensity' : output[6],
                                 'mean' : output[7],
                                 'standard_deviation' : output[8],
                                 'lamda_max_min_diff' : str(float(output[3]) - float(output[4])),
                                 'answer' : answer 
                                 })
    return path_to_csv
```

```python
def mix_positive_and_negative_to_csv(positive_data_frame, negative_data_frame, output_file):
    frames = [positive_data_frame, negative_data_frame]
    data_frame = pd.concat(frames,ignore_index=True)
    data_frame = data_frame.reindex(np.random.permutation(data_frame.index))
    data_frame.to_csv(output_file)
    return data_frame
```

```python
def stamp_classifier(data, threshold):
    answers = []
    for i in data:
        if i > threshold:
            answers.append(1)
        else:
            answers.append(0)
    return answers
```

```python
def choose_the_best_threshold(min_val, max_val, step, real_answers, data):
    
    best_threshold = min_val
    min_err = len(data)
    best_classifier_answers = [-1 for i in range(len(data))]
    precision = []
    recall = []
    TPR = []
    FPR = []
    TNR = []
    for threshold in np.arange(min_val, max_val, step):
        classifier_answers = stamp_classifier(data, threshold)
        
        err, pre, rec, tpr, fpr, tnr = error(real_answers, classifier_answers)
        precision.append(pre)
        recall.append(rec)
        TPR.append(tpr)
        FPR.append(fpr)
        TNR.append(tnr)
        if err < min_err:
            best_threshold = threshold
            min_err = err
            best_classifier_answers = classifier_answers
            print ('min_err = ', min_err, 'th',  best_threshold)
#     disc_func = disc_func_of_simpe_lamda_diff_classifier(data, threshold)

    plot_graph(recall, precision, 'recall', 'precision') 
    plot_graph(FPR, TPR,'FPR', 'TPR')
   
    return best_threshold, best_classifier_answers#, disc_func
```

```python
def plot_graph(x, y, x_label, y_label):
    plt.plot(x, y, linestyle='-', marker='o', color='blue')
    plt.xlabel(x_label, fontsize=10)
    plt.ylabel(y_label, fontsize=10)
    plt.show()
```

```python
def error(real_answers, classifier_answers):
    error = 0.0
    pre = 0.0
    rec = 0.0
    
    tp = 0.0
    fp = 0.0
    fn = 0.0
    tn = 0.0
    for i in range(len(real_answers)):
        if (real_answers[i] != classifier_answers[i]):
            error += 1.0
        if ((real_answers[i] == 1) and (classifier_answers[i] == 1)):
            tp += 1.0
        if ((real_answers[i] == 0) and (classifier_answers[i] == 1)):
            fp += 1.0
        if ((real_answers[i] == 1) and (classifier_answers[i] == 0)):
            fn += 1.0
        if ((real_answers[i] == 0) and (classifier_answers[i] == 0)):
            tn += 1.0
            
    pre = tp / (tp + fp)
    rec = tp / (tp + fn)
    tpr = tp / (tp + fn)
    tnr = tn / (tn + fp)
    fpr = fp / (fp + tn)

    error /= len(real_answers)
    return error, pre, rec, tpr, fpr, tnr
```

```python
def plot_precision_recall_curve(data_frame, threshold):
    new_data_frame = data_frame
#     lamda_max = new_data_frame['lamda_max'].values.tolist()
    lamda_min = new_data_frame['lamda_min'].values.tolist()
    
    disc_func = []
    for i in range(len(lamda_min)):
        disc_func.append(lamda_min[i])#(lamda_max[i] - lamda_min[i])
    new_data_frame['disc_func'] = disc_func #/ max(np.array(disc_func))
    new_data_frame['classifier_answers'] = stamp_classifier(disc_func, threshold)    
    
    sorted_data_frame = new_data_frame.sort(['disc_func'], ascending=False)
    y_true = sorted_data_frame['answer'].values.tolist()
    y_scores = sorted_data_frame['disc_func'].values.tolist()
    precision, recall, thresholds = precision_recall_curve(y_true, y_scores)
    
    print('standart recall-precision curve')
    plot_graph(recall, precision, 'recall', 'precision')
    
    print('standart roc curve')
    fpr, tpr, _ = roc_curve(y_true, y_scores)
    plot_graph(fpr, tpr,'FPR', 'TPR')
```
